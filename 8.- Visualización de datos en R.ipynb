## Grafico 1 : Peso vs precio (¿A mayor peso mayor precio?)

```r
# Gráfico de puntos
p1 <- ggplot(diamonds, aes(x=carat, y=price)) +
  # definimos la geometría (gráfico de puntos)
  geom_point() +
  # título del eje x
  xlab('Carat') +
  # título del eje y
  ylab('Price') +
  # titulo principal
  ggtitle('Carat vs price')

p1 
```

## Relación peso vs precio visto por corte

```r
# Mismo gráfico pero identificando el corte por color
p2 <- ggplot(diamonds, aes(x=carat, y=price, color=cut)) +
  # definimos la geometría (gráfico de puntos)
  geom_point() +
  # título del eje x
  xlab('Carat') +
  # título del eje y
  ylab('Price') +
  # titulo principal
  ggtitle('Carat vs price') +
  # Se modifica el tamaño de fuente  de leyenda de ejes y titulo
  theme(axis.text=element_text(size=12), # modifica el tamaño de la fuente de los ejes x, y
        axis.title=element_text(size=20), # modifica el tamaño de la fuente de eiquetas de los ejes x, y
        plot.title = element_text(hjust = 0, size=24, color="blue"), # modifica el posición, tamaño y color del título
        legend.position="bottom", # mueve la posición de la leyenda a la parte inferior
        legend.box = "horizontal") # Si existen muchas legendas las pone de forma horizantal

p2 
```

## Relación peso vs precio visto por claridad

```r
# Mismo gráfico pero identificando la claridad por color
# definimos nuestros colores 
col <- c("#999999", "#E69F00", "#56B4E9", "#009E73", "#F0E442", "#0072B2", "#D55E00", "#CC79A7")
# graficamos
# Mismo gráfico pero identificando el corte por color
p3 <- ggplot(diamonds, aes(x=carat, y=price, color=clarity)) +
  # definimos la geometría (gráfico de puntos)
  geom_point() + 
  # título del eje x
  xlab('Carat') +
  # título del eje y
  ylab('Price') +
  # cambio de limite de eje x
  xlim(0,5.2) +
  # titulo principal
  ggtitle('Carat vs price') +
  # Se modifica el tamaño de fuente  de leyenda de ejes y titulo
  theme(axis.text=element_text(size=12), # modifica el tamaño de la fuente de los ejes x, y
        axis.title=element_text(size=20),# modifica el tamaño de la fuente de eiquetas de los ejes x, y
        plot.title = element_text(hjust = 0, size=24, color="blue"), # modifica el posición, tamaño y color del título
        legend.position="bottom",    # mueve la posición de la leyenda a la parte inferior
        legend.box = "horizontal") + # Si existen muchas legendas las pone de forma horizantal
  # colores definidos manualmente
  scale_colour_manual(values=col)

p3 
```

#  FACETS

## Relación peso vs precio visto por corte (un gráfico por corte)

```r
# facet_wrap (separa gráficos en subconjuntos)
p4 <- ggplot(diamonds, aes(x=carat, y=price, color=cut)) +
  # definimos la geometría (gráfico de puntos)
  geom_point() +
  # recta suavizada
  geom_smooth() +
  # título del eje x
  xlab('Carat') +
  # título del eje y
  ylab('Price') +
  # cambio de limite de eje x
  xlim(0,5.2) +
  # titulo principal
  ggtitle('Carat vs price') +
  # Se modifica el tamaño de fuente  de leyenda de ejes y titulo
  theme(axis.text=element_text(size=12), # modifica el tamaño de la fuente de los ejes x, y
        axis.title=element_text(size=20), # modifica el tamaño de la fuente de eiquetas de los ejes x, y
        plot.title = element_text(hjust = 0, size=24, color="blue"), # modifica el posición, tamaño y color del título
        legend.position="bottom",    # mueve la posición de la leyenda a la parte inferior
        legend.box = "horizontal") + # Si existen muchas legendas las pone de forma horizantal
  # separa un gráfico por la variable corte
  facet_wrap(~cut) 
p4
```

# Relación peso vs precio visto por corte y claridad (un gráfico por corte y claridad)

```r
# Facet grid
p5 <- ggplot(diamonds, aes(x=carat, y=price, color=cut)) +
  # definimos la geometría (gráfico de puntos)
  geom_point() +
  # título del eje x
  xlab('Carat') +
  # título del eje y
  ylab('Price') +
  # cambio de limite de eje x
  xlim(0,5.2) +
  # titulo principal
  ggtitle('Carat vs price') +
  # Se modifica el tamaño de fuente  de leyenda de ejes y titulo
  theme(axis.text=element_text(size=12), # modifica el tamaño de la fuente de los ejes x, y
        axis.title=element_text(size=20),# modifica el tamaño de la fuente de eiquetas de los ejes x, y
        plot.title = element_text(hjust = 0, size=24, color="blue"), # modifica el posición, tamaño y color del título
        legend.position="bottom",    # mueve la posición de la leyenda a la parte inferior
        legend.box = "horizontal") + # Si existen muchas legendas las pone de forma horizantal
  # separa un gráfico por la variable corte
  facet_grid(cols=vars(cut), rows= vars(clarity), scales="free")

p5
```

# Tipos de geometrías

|Name | Description |
|----|----|
| abline | Line, specified by slope and intercept |
| area   | Area plots |
| bar    | Bars, rectangles with bases on y-axis |
| blank  | Blank, draws nothing |
| boxplot | Box-and-whisker plot |
| contour | Display contours of a 3d surface in 2d |
| crossbar | Hollow bar with middle indicated by horizontal line |
| density | Display a smooth density estimate |
| density 2d | Contours from a 2d density estimate |
| errorbar | Error bars |
| histogram | Histogram |
| hline | Line, horizontal |
| interval | Base for all interval (range) geoms |
| jitter | Points, jittered to reduce overplotting |
| line  | Connect observations, in order of x value |
| linerange | An interval represented by a vertical line
| path | Connect observations, in original order |
| point | Points, as for a scatterplot |
| pointrange | An interval represented by a vertical line, with a point in the middle |
| polygon | Polygon, a filled path |
| quantile | Add quantile lines from a quantile regression |
| ribbon | Ribbons, y range with continuous x values |
| rug | Marginal rug plots |
| segment | Single line segments |
| smooth | Add a smoothed condition mean |
| step | Connect observations by stairs |
| text | Textual annotations |
| tile | Tile plot as densely as possible, assuming that every tile  is the same size |
| vline | Line, vertical |

## Histograma

## ¿Cómo se distribuye el precio?

```r
p7 <- ggplot(diamonds, aes(x=price)) +
    geom_histogram()
p7
```

## Distribución del precio por claridad (un gráfico por claridad)

```r
# histograma
p8 <- ggplot(diamonds, aes(x=price)) +
    geom_histogram(aes(y=..density.., fill=clarity), alpha=0.7, bins = 30) +
    geom_density(aes(fill=clarity, color=clarity), alpha=0.2) +
    facet_wrap(~clarity, scales="free") +
    theme(axis.text=element_text(size=12), # modifica el tamaño de la fuente de los ejes x, y
          axis.title=element_text(size=20), # modifica el tamaño de la fuente de eiquetas de los ejes x, y
          plot.title = element_text(hjust = 0, size=24, color="blue"), # modifica el posición, tamaño y color del título
          legend.position="bottom",    # mueve la posición de la leyenda a la parte inferior
          legend.box = "horizontal")
p8
```

# Gráfico de puntos

## Grálfico de el promedio de x contra el promedio del promedio del precio (price) por la variable corte (cut)

```r
df01 <- diamonds  %>% 
    group_by(cut)  %>% 
    summarize(mean_x = mean(x), mean_price = mean(price))  %>% 
    ungroup()  %>% 
    arrange(desc(mean_price))

df01    
```

```r
# gráfico de puntos
p9 <- ggplot(df01, aes(x=mean_x, y=mean_price, color=cut)) +
    geom_point() +
    geom_text(aes(label=cut, color=cut))
p9
```
