# **If else**

A menudo es útil forzar la ejecución de alguna parte de un programa para que dependa de una condición. La función $if$ tiene la forma

```r
if (logical_expression) {
  expression_1
  ...
}
```

Una extensión natural del comando $if$ incluye la parte $else$
```r
if (logical_expression) {
  expression_1
  ...
} else {
  expression_2
  ...
}
```

Cuando se evalúa una expresión $if$, si la expresion logical es TRUE se ejecuta la expresión del primer grupo y no se ejecuta la del segundo grupo. A la inversa, si la expresión lógica es FALSE, sólo se ejecuta la expresión del segundo grupo. 

**Estructura de selección múltiple.**

En ocasiones resulta necesario evaluar múltiples condiciones, de la siguiente forma 

```r
if (logical_expression_1) {
  expression_1
  ...
} else if (logical_expression_2) {
  expression_2
  ...
} else {
  expression_3
  ...
} else if (logical_expression_2) {
  expression_4
  ...
} else {
  expression_5
  ...
}
```

# **Bucle While**

A menudo no sabemos de antemano cuántas veces tenemos que dar la vuelta a un bucle. Es decir, cada vez que damos una vuelta al bucle, comprobamos alguna condición para para ver si ya hemos terminado. En esta situación utilizamos un bucle $while$, que tiene la forma

```r
while (logical_expression) {
    expression_1
    ...
}
```

Cuando se ejecuta un comando $while$, se evalúa primero la expresión lógica. Si es **TRUE** se ejecuta el grupo de expresiones entre llaves { }. A continuación, el control se vuelve al inicio del comando: si la expresión lógica sigue siendo **TRUE**, las expresiones agrupadas se ejecutan de nuevo, y así sucesivamente. Evidentemente, para que el bucle se detenga finalmente, la expression lógica debe ser finalmente **FALSE**. Para conseguirlo, la expression lógica suele depender de una variable que se altera dentro de las expresiones agrupadas.

El bucle $while$ es más fundamental que el bucle $for$, ya que siempre podemos reescribir un bucle $for$ como un bucle $while$.

## Ejemplo 1:

Supongamos que queremos conocer el primer número entero positivo cuyo cuadrado es superior a $4000$.

````r
# Variable initialization
n <- 0
square <- 0
# While loop
while(TRUE) {
    n <- n + 1
    square <- n ^ 2
    if(square > 4000){
      break # palabra reservada rompe el ciclo
    }
}
print(n)
print(square)
````

El usuario tecleará los votos de manera desorganizada, tal y como se obtuvieron en la elección, el final de datos está representado por un cero. Observe, como ejemplo, la siguiente lista:

$$1,  3,  1,  4,  2,  2,  1,  4,  1,  1,  1,  2,  1,  3,  1,  4,  0$$

Donde 1 representa un voto para el candidato 1; 3 un voto para el candidato 3;

````r
# Solcución aquí
x <- c(1, 3, 1, 4, 2, 2, 1, 4, 1, 1, 1, 2, 1, 3, 1, 4, 0)
# creamos variable que cuentan votos de candidatos
votos_c1 <- 0
votos_c2 <- 0
votos_c3 <- 0
votos_c4 <- 0
i <- 1
# ciclo while
while(x[i] != 0){
  if(x[i] == 1){
    votos_c1 <- votos_c1 + 1
  } else if(x[i] == 2){
    votos_c2 <- votos_c2 + 1
  } else if(x[i] == 3){
    votos_c3 <- votos_c3 + 1
  } else if(x[i] == 4){
    votos_c4 <- votos_c4 + 1
  }
  i <- i + 1
}
cat('Votos candidato 1 :', votos_c1, '\n')
cat('Votos candidato 2 :', votos_c2, '\n')
cat('Votos candidato 3 :', votos_c3, '\n')
cat('Votos candidato 4 :', votos_c4, '\n')
````

#**Observación:**

Un ciclo **while** es más fundamental que un ciclo **for**, eso quiere decir que todo ciclo **for** puede codificarse mediente un ciclo **while**, pero no a la inversa.

Pensemos en la siguiente situación:
Supongamos que vamos a apostar en lanzamientos sucesivos de monedas. Si cae cara ganamos $1$ y si cae cruz perdemos $1$. Analicemos la situación de dos jugadores:

- Un jugador decide apostar únicamente $10$ jugadas, gane o pierda.
- Otro jugador decide apostar hasta que pierda todo. 

En el primer caso podríamos utilizar un la sentencia **for** puesto que conocemos el total de jugadas, sin embargo, en el segundo caso no conocemos el número total de jugadas, y pude suceder que el jugador, del segundo caso, pierda todo en $100$ jugadas, en $1000$, o en otro número mayor, en tal situación no podemos emplear la sentencia **for** para simular las sucesivas apuestas, pero podríamos emplear la sentencia **while**.

## Ejemplo 8: Caso 1: Apuesta 10 jugadas

Generar un código para calcular una realización de 

\begin{equation}
S_n = S_{n-1} + X_i, \ \ \ \ S_0 = 10
\end{equation}

donde $P(X_i=1) = p$ , y $P(X_i=-1) = 1-p$. Con la condición de que acumule hasta que $S_n=0$.

````r
sample(c(1, -1), size=1, replace=T, prob=c(0.5, 0.5))
````

````r
# probabilidad de ganar
p <- 0.5
# monto inicial
S0 <- 10
# n jugadas 
n <- 10
# valores que puede tomar 
x.gana = 1 # a la derecha
x.pierde = -1 # a la izquierda
# inicialización del vector
Sn = c(S0)
# inicio del bucle
for(i in 2:(n+1)){
    x_i <- sample(c(x.gana, x.pierde), size=1, replace=T, prob=c(p, 1-p))
    Sn[i]  <- Sn[i-1] + x_i
 }
 # impresión del resultado
 print(Sn)
````

donde $P(X_i=1) = p$ , y $P(X_i=-1) = 1-p$. Con la condición de que acumule hasta que $S_n=0$.

````r
# probabilidad de ganar
p <- 0.5
# monto inicial
S0 <- 10
# valores que puede tomar 
x.gana = 1 # a la derecha
x.pierde = -1 # a la izquierda
# inicialización del vector
Sn = c(S0)
# inicialización
i <- 2
# inicio del bucle
while(Sn[i-1] != 0){
    x_i <- sample(c(x.gana, x.pierde), size=1, replace=T, prob=c(p, 1-p))
    Sn[i]  <- Sn[i-1] + x_i
    i <- i + 1
 }
 # impresión del resultado
 print(Sn)
 print(length(Sn))
````

````r
df <- data.frame('Jugadas'=0:(length(Sn)-1), 'Capital'=Sn)
````
