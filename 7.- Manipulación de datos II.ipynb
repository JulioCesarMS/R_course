# **group_by() y summarize()**

> **group_by(data, ...)** Sirve para agrupar filas (filas) de un data frame para obtener variables de resumen.

**Argumentos:**

- **data** : dataframe
- ...  : variable o grupo de variables a ser agrupadas.

groupby.png

 DespuÃ©s de aplicar la funciÃ³n group_by() es necesario aplicar una funciÃ³n resumen como summarize().

> **summarize()** : Sirve para obtener resumen de variables, dentro de esta podemos especifÃ­car funciones resumen como : 
> - sum()
> - mean() 
> - sd()
> - var()
> - max()
> - min()
> - n()
> - n_distinct()
> - first()
> - last()


> **ungroup()** : sirve para desagrupar, es conveniente desagrupar cuando se requiera hacer calculos con valores de variables desagrupados.

# **pivot_longer()**

> *pivot_longer(data, cols, names_to = "name", values_to = "value", ...)*

- **data** : dataframe 
- **cols** : columnas a formato long
- **names_to** : Una cadena que especifica la nueva columna.
- **values_to : Una cadena que especifica el nombre de la columna a crear a partir de los datos almacenados en los valores de las celdas.

# **pivot_wider()**

> *pivot_wider(data, names_from = "name", values_from = "value", values_fill, ...)*

- **data** : dataframe

# **Joins** 

Nos permite combinar filas de dos o mÃ¡s tablas  basÃ¡ndose en una o varias columnas relacionadas entre ellas.

- **inner_join(x, y, by, suffix = c(".x", ".y"), ...)**: incluye todas las filas de $x$ y $y$.

- **left_join(x, y, by, suffix = c(".x", ".y"), ...)**: incluye todas las filas de $x$.

- **right_join(x, y, by, suffix = c(".x", ".y"), ...)**: incluye todas las filas de $y$.

- **full_join(x, y, by, suffix = c(".x", ".y"), ...)**: incluye todas las filas de x o y.

```r
df_delitos_ll <- df_delitos %>%
  left_join(ag, by=c('Entidad'='Nom_Ent', 'Municipio'='Nom_Mun')) %>%
  #merge(ag, by.x=c('Entidad', 'Municipio'), by.y=c('Nom_Ent', 'Nom_Mun'))
  pivot_longer(Enero:Diciembre, names_to='Mes', values_to='Valor') %>% 
  left_join(mes, by='Mes') %>% 
  mutate('Fecha' = as.Date(paste(AÃ±o, MesNum,'01', sep='-'))) %>%
  select(Fecha, everything()) %>%
  filter(`Tipo de delito`=='Homicidio')

head(df_delitos_ll)
```

```r
df_map <- df_delitos_ll 
```

```r
#install.packages("leaflet")
#install.packages("htmlwidgets")
#install.packages("IRdisplay")
library("leaflet")
library("htmlwidgets")
library("IRdisplay")
```

```r
year = 2019
```

```r
df_world <- df_map %>%
    group_by(AÃ±o, Entidad, `Tipo de delito`) %>%
    summarise('Valor' = sum(Valor),
              'Lat' = mean(Lat),
              'Lon' = mean(Lon)) %>%
    ungroup() %>%
    mutate(radio = ifelse(Valor<=100,1,
                          ifelse(Valor<=300,5,
                                 ifelse(Valor<=500,10,
                                        ifelse(Valor>500,20,0))))) %>%
    filter(AÃ±o==year) %>%
    #filter(Valor >0) %>%
    filter(!is.na(Lon), !is.na(Lat))
  
```

```r
plt <- leaflet(df_world) %>% 
    addTiles() %>% 
    addCircleMarkers(
      #clusterOptions = markerClusterOptions(),
      lng = df_world$Lon,
      lat = df_world$Lat,
      radius = df_world$radio,
      color = 'red',
      popup = paste('<b><a>AÃ±o :</a></b> ',df_world$AÃ±o,"<br/>" ,
                     '<b><a>Entidad :</a></b> ',df_world$Entidad,"<br/>" ,
                     '<b><a>Homicidios Totales:</a></b> ',df_world$Valor,"<br/>")) %>%
    setView(lng = mean(df_world$Lon),lat = mean(df_world$Lat), zoom=5)%>%
    addProviderTiles("Stadia.AlidadeSmoothDark") # Stadia.AlidadeSmoothDark
```

```r
saveWidget(plt, 'map.html', selfcontained = TRUE)
```
